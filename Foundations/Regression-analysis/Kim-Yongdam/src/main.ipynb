## Proximity

### L2 distance(Euclidean distance)

```python
def euclidean(a, b):
    return np.sqrt(
        np.sum(
            np.square(a - b) # Element-wise
        )
    )
```

```python
f = lambda x, y: np.sqrt(np.sum(np.square(x, y)))
f(v1, v2)
```

### Cosine similarity

```python
def cosine_sim(a, b):
    # Numerator: a o b
    dot = a @ b
    
    # Denominator: ||a||, ||b||
    norm = np.linalg.norm(a) * np.linalg.norm(b) + 1e-12 # Prevent Divided by 0
    
    sim = dot / norm
    if np.isnan(sim):
        return 0
    else:
        return sim
```

```python
print(euclidean(v1, v2))
cosine_sim(v1, v2)
```

```python
dist = np.sqrt(np.sum(np.square(v1 - v2)))
dist2 = np.linalg.norm(v1 - v2)
print(dist, dist == dist2)
```
